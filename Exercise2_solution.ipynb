```python
trainedit = pd.read_csv('train edited.csv')
```

```python
trainedit.SalePrice.describe(include = [np.number])
```

```python
f = plt.figure(figsize=(24, 4))
sb.boxplot(data = price, orient = 'h')
```

```python
f = plt.figure(figsize=(16, 8))
sb.histplot(data = price)
```

```python
f = plt.figure(figsize= (16, 8))
sb.histplot(data = price, kde = True)
```

```python
lot = pd.DataFrame(trainedit['LotArea'])
```

```python
lot.describe(include = 'all')
```

```python
f = plt.figure(figsize= (24,4))
sb.boxplot(data = lot, orient = 'h')
f = plt.figure(figsize= (14,4))
sb.histplot(data = lot, kde = True)
```

```python
joint = pd.concat([price, lot], axis = 1).reindex(price.index)
print(joint)
```

```python
sb.jointplot(data = joint, x = "SalePrice", y = "LotArea", height = 12)
```
