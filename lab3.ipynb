```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
sb.set()
data = pd.read_csv('train.csv')
houseNumData = pd.DataFrame(data[['LotArea', 'GrLivArea', 'TotalBsmtSF', 'GarageArea', 'SalePrice']])
for colum in houseNumData:
    print(colum)
    plt.figure()
    sb.histplot(data = data[colum], kde = True)
```

```python
for colum in houseNumData:
    print(colum)
    print(houseNumData[colum].skew())
```

```python
f = plt.figure(figsize=(12,12))
sb.heatmap(houseNumData.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

```python
sb.pairplot(data = houseNumData)
```

```python
for col in houseNumData:
    if col == 'SalePrice':
        continue
    sb.jointplot(data = houseNumData, x = col, y = "SalePrice")
    
```

```python
houseCatData = pd.DataFrame(data[['MSSubClass', 'Neighborhood', 'BldgType', 'OverallQual', 'SalePrice']])
for cat in houseCatData:
    sb.catplot(data = houseCatData, x = cat, y = 'SalePrice')
    
```
